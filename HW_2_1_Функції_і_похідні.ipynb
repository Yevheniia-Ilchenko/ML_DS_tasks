```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')

f_lin = 10 * x - 5
derivative_f_lin = sp.diff(f_lin, x)
print("Похідна:", derivative_f_lin)
```

```python
def f(x):
  return 10 * x-5

def f_derivative(x):
  return 10
```

```python
x = np.linspace(0, 10, 100)

y = f(x)
y_derivative = np.full_like(x, 10)

plt.figure(figsize=(10, 5))

plt.plot(x, y, label='f(x) = 10x−5', color='blue')
plt.plot(x, y_derivative, label="f'(x) = 10", color='green')


plt.xlabel("Hours")
plt.ylabel("Graude")
```
